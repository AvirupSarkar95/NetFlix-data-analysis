```python
# Generating WordCloud of the 200 most used words
from collections import Counter
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
decription=list(nfx['description'])
dsp=[]
```

```python
for i in decription :
    i=list(i.split(','))
    for j in i :
        dsp.append(j.replace(' ',""))
d_count=Counter(dsp)    
```

```python
txt=list(set(dsp))
plt.rcParams['figure.figsize']=(20,20)
wordcloud=WordCloud(max_font_size=50,max_words=200,background_color="black").generate(str(txt))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
